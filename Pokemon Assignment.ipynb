```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv("/Users/Ada/Desktop/DS5500/Assignment/Visual Encoding with Matplotlib/Pokemon.csv")
pokemon
```

```python
pokemon_performance_by_generation = pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
pokemon_performance_by_generation.plot.line()
plt.title("Polemon Performance By Generation")

```

The analysis is focused on the generation. It is a low-level action.

```python
pokemon_performance_by_generation = pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
pokemon_performance_by_generation.plot.line()
plt.title("Pokemon Performance By Generation")
```

The question we are trying to solve here is what is the difference of pokemon performance between the generations.  The encoding I didn't use is the pie plot (make a pie plot for each generation).Categorical attributes should be shown with the identity channels. The most effective channel for categorical data is spatial region with color hue. Area is not a relatively effective or intuitive way to present the information. (Book Page 101)
The advantage of the multi-line plot is that it gives accurate comparsion of data without overdressing. Like we discussed in class, the area can easily give false feeling of the proportion. Line is the most effective way to demonstrate such information. 


```python
pokemon_stats_legendary = pokemon.groupby(['Legendary', 'Generation']).mean()[['Attack', 'Defense']]
pokemon_stats_legendary.plot.bar(stacked=False)
plt.title("Pokemon Performance By Generation and Legendary")
```

The question we are trying to solve here is what is the difference of normal pokemon and legendary pokemon performance between the generations. 
The encoding I didn't use is the stacked bar plot. According to the effectiveness principle: the importance of the attribute should match the salience of the channel. We need to encode most important attributes with highest ranked channels (Lecture 3 Page 41). In our case, it is the difference between the legendary and normal pokemon among generations so it is more intuitive to apply dodge bar plot. We can now see that the legendary pokemon have higher attack and defense for all generations.




I prefer the first visualization because it is very straightforward.
